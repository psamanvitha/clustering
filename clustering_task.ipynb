```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = data[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
```

```python
fig, ax = plt.subplots()
ax.scatter(X['sepal_length'], X['sepal_width'], c=kmeans.labels_, cmap='viridis')
ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_width')
plt.show()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=5, random_state=32)
kmeans.fit(X)
fig, ax = plt.subplots()
ax.scatter(X['petal_length'], X['petal_width'], c=kmeans.labels_, cmap='viridis')
ax.set_xlabel('petal_length (cm)')
ax.set_ylabel('petal_width (cm)')
plt.show()
```
